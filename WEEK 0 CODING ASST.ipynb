```python
train = 'train.csv'
df_food = pd.read_csv(train)
df_food.head()
```

```python
df_merged = pd.merge(df_meal,df_food, on='meal_id')

df_merged.head()

table = pd.pivot_table(df_merged,index='category',values='num_orders',aggfunc=np.sum)
print(table)
```

```python
table.plot(kind='bar')
plt.xticks(rotation=70)
plt.xlabel('Food Items')
plt.ylabel('Quantity sold')
plt.title('Most popular food')
plt.show()
```
